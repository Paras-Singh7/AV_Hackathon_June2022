```python
train['created_at'] = pd.to_datetime(train['created_at'])
train.signup_date = pd.to_datetime(train.signup_date)
test['created_at'] = pd.to_datetime(test['created_at'])
test.signup_date = pd.to_datetime(test.signup_date)
```

```python
train['created_day'] = train.created_at.apply(lambda x: x.day)
train['created_month'] = train.created_at.apply(lambda x: x.month)
train['created_year']= train.created_at.apply(lambda x: x.year)
"""train['signup_day']= train.signup_date.apply(lambda x: x.day)
train['signup_month'] = train.signup_date.apply(lambda x: x.month)
train['signup_year'] = train.signup_date.apply(lambda x: x.year)"""
train['cday_of_week'] = train.created_at.apply(lambda x: x.dayofweek)
train['cis_quarter'] = train.created_at.apply(lambda x: x.quarter)
train['cis_weekend'] = train.cday_of_week.apply(lambda x: 1 if x>4 else 0)
train.drop(['created_at', 'signup_date'], axis=1, inplace=True)
train.head()
```

```python
test['created_day'] = test.created_at.apply(lambda x: x.day)
test['created_month'] = test.created_at.apply(lambda x: x.month)
test['created_year']= test.created_at.apply(lambda x: x.year)
"""test['signup_day']= test.signup_date.apply(lambda x: x.day)
test['signup_month'] = test.signup_date.apply(lambda x: x.month)
test['signup_year'] = test.signup_date.apply(lambda x: x.year)"""
test['cday_of_week'] = test.created_at.apply(lambda x: x.dayofweek)
test['cis_quarter'] = test.created_at.apply(lambda x: x.quarter)
test['cis_weekend'] = test.cday_of_week.apply(lambda x: 1 if x>4 else 0)
test.drop(['created_at', 'signup_date'], axis=1, inplace=True)
test.head()
```

```python
train = train.fillna(0)
test = test.fillna(0)
```

```python
train['activity'] = train.iloc[:,3:15].sum(axis = 1)
train['active_user'] = train.activity.apply(lambda x: 1 if x>=1 else 0 )
```

```python
test['activity'] = test.iloc[:,3:15].sum(axis = 1)
test['active_user'] = test.activity.apply(lambda x: 1 if x>=1 else 0 )
```

```python
y = train['buy'].values
x = train.drop('buy', axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import  mean_squared_error
import lightgbm as lgb
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.01, random_state=41)
```

```python
classifier = RandomForestClassifier( random_state=123, max_depth=15, n_estimators=1000,min_samples_leaf=4, min_samples_split=2)
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
confusion_matrix(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
gbm = lgb.LGBMClassifier(num_leaves=89,learning_rate=0.1,max_bin=200,min_sum_hessian_in_leaf = 20,
                        max_depth= 18,verbose=-1)
```

```python
gbm.fit(x_train, y_train,eval_set=[(x_test, y_test)], eval_metric='l2_root', early_stopping_rounds=100)
```

```python
y_pred = gbm.predict(x_test, num_iteration=gbm.best_iteration_)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```
